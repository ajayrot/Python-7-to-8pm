# Turtle Methods:
- Turtle 
- forward()
- backward()
- right()
- left()
- penup()
- pendown()
- up()
- down()
- color()

```python
# Drwa a triangle
import turtle as t
turtle = t.Turtle()

for i in range(3):
    t.forward(100) # length
    t.right(120) # Angle
t.done()
```

```python
# drwa sqaure
turtle = t.Turtle()
for i in range(4):
    t.backward(100)
    t.left(90)
t.done()
```

```python
# Draw a pentagon
turtle = t.Turtle()
for i in range(5):
    t.backward(100)
    t.left(72)
t.done()
```

```python
# Spiraling star
turtle = t.Turtle()
t.pencolor("red")
for i in range(50):
    t.forward(i*10)
    t.right(144)
t.done()
```

```python
# sqaure spiraling
turtle = t.Turtle()
t.pencolor("blue")
for i in range(150):
    t.forward(i)
    t.right(91)
t.done()
```

```python
# Hexagon with multiple colors
from turtle import *
colors = ['blue','red','yellow','green','orange','purple']
for i in range(360):
    pencolor(colors[i%6])
    width((i/100+1))
    speed(100)
    forward(i)
    left(59)
```
